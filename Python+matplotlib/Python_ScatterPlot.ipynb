```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

```python
df = pd.read_csv('C:/Users/rin/Desktop/02-DataVis-5Ways-1/cars-sample.csv')
```

```python
# fig = plt.subplot()
fig,ax = plt.subplots()
for index, row in df.iterrows():
    x = row['Weight']
    y = row['MPG']
    m = row['Manufacturer']
    area = x/25
    color = ''
    if m == 'bmw':
        color = 'pink'
    elif m == 'ford':
        color = 'green'
    elif m == 'honda':
        color = 'cyan'
    elif m == 'mercedes':
        color = 'orange'
    elif m == 'toyota':
        color = 'purple'
    ax.scatter(x, y, s=area, c=color, alpha=0.5)

legend_color = [Line2D([0], [0], marker='o', color='w', label='bmw',
                          markerfacecolor='pink', markersize=5),
             Line2D([0], [0], marker='o', color='w', label='ford',
                          markerfacecolor='green', markersize=5),
             Line2D([0], [0], marker='o', color='w', label='honda',
                          markerfacecolor='cyan', markersize=5),
             Line2D([0], [0], marker='o', color='w', label='mercedes',
                          markerfacecolor='orange', markersize=5),
             Line2D([0], [0], marker='o', color='w', label='toyota',
                          markerfacecolor='purple', markersize=5),
             Line2D([0], [0], marker='o', color='w', label='2000',
                          markerfacecolor='grey', markersize=8),
             Line2D([0], [0], marker='o', color='w', label='3000',
                          markerfacecolor='grey', markersize=10),
             Line2D([0], [0], marker='o', color='w', label='4000',
                          markerfacecolor='grey', markersize=12)]


ax.legend(handles=legend_color, loc='best', bbox_to_anchor=(1.3, 1))
plt.xlabel('Weight')
plt.ylabel('MPG')


plt.show()
```
