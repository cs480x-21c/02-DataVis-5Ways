```python
import matplotlib.pyplot as plt
import csv
import pandas as pd

df = pd.read_csv('cars-sample.csv')
df[['Weight']] = df[['Weight']].apply(pd.to_numeric) / 30
grouped = df.groupby(df.Manufacturer)

#print(grouped.get_group('bmw'))

manufacturers = {'bmw' : 'orange',
                 'ford' : 'olive',
                 'honda' : 'teal',
                 'mercedes' : 'aqua',
                 'toyota' : 'fuchsia'}
for brand in manufacturers:
    curr_brand = grouped.get_group(brand)
    plt.scatter(data=curr_brand, x='Weight', y='MPG', alpha=0.5, s='Weight', c=manufacturers.get(brand))

plt.xlabel('Weight', size=16)
plt.ylabel('MPG', size=16)
# bmw = grouped.get_group('bmw')
# ford = grouped.get_group('ford')
# honda = grouped.get_group('honda')
# mercedes = grouped.get_group('mercedes')
# toyota = grouped.get_group('toyota')

# plt.scatter(data=bmw, x='Weight', y='MPG', alpha=0.5, s='Weight', c='blue')

```
