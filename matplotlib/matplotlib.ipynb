```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator, NullFormatter)

# Color dictionary
colors = {
    'bmw': '#0DC800', 
    'ford': '#FF8C00', 
    'honda': '#11B1DD', 
    'mercedes': '#4800FF', 
    'toyota': '#FF006F'
    }

# Read in csv
data = pd.read_csv('../cars-sample.csv')

# Plot points
plt.scatter(data.Weight, data.MPG, s=(data.Weight/400)**2, c=data.Manufacturer.map(colors), alpha=0.5, edgecolors='none')

# Axis labels
plt.xlabel("Weight")
plt.ylabel("MPG")

# Customize ticks
plt.xticks(np.arange(2000, 6000, 1000))
plt.yticks(np.arange(10, 50, 10))

# Set background color 
ax = plt.gca()
ax.set_facecolor("#ececec")

# Set gridlines
ax.set_axisbelow(True)
ax.grid(which='minor', color='white', linestyle='-', linewidth=0.5)
ax.grid(which='major', color='w', linestyle='-', linewidth=1)
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))

# Eliminate all tick marks
for tic in ax.xaxis.get_major_ticks():
    tic.tick1On = tic.tick2On = False
for tic in ax.xaxis.get_minor_ticks():
    tic.tick1On = tic.tick2On = False
for tic in ax.yaxis.get_major_ticks():
    tic.tick1On = tic.tick2On = False
for tic in ax.yaxis.get_minor_ticks():
    tic.tick1On = tic.tick2On = False

plt.show()
```
