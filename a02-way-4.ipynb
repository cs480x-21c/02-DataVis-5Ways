```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv("cars-sample.csv")
```

```python
cars = cars.sort_values('Manufacturer', ascending=True)
```

```python
cars
```

```python
fig, ax = plt.subplots(figsize=(8.5,6), dpi=120)
sns.set(style='darkgrid')
sns.scatterplot(data=cars, x='Weight', y='MPG', hue='Manufacturer', size='Weight', alpha=0.5, palette='husl',
               sizes=(20, 200))
ax.set_yticks([10,20,30,40])
ax.set_xticks([2000,3000,4000,5000])
plt.legend(bbox_to_anchor=(1.02, .75), loc=2, borderaxespad=0., labelspacing=.7,)
plt.tight_layout()
plt.savefig("way-4-matplotlib-seaborn.png")
plt.show()
```
